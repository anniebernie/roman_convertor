```python
# DEF FUNCTIONS


# =========================================================================
def decomp(base_number, base_numbers, factors, order, roman_base_numbers):
  cent_part = ""
  base_number = base_number
  base_number_index = base_numbers.index(base_number)
  factor_centaine = factors[order]

  first_elemt = [base_numbers[base_number_index], roman_base_numbers[base_number_index]]
  second_elemt = [base_numbers[base_number_index+1], roman_base_numbers[base_number_index+1]]
  third_elemt = [base_numbers[base_number_index+2], roman_base_numbers[base_number_index+2]]

  # print(first_elemt)
  # print(second_elemt)
  # print(third_elemt)

  if factor_centaine < 4:
    cent_part = factor_centaine*first_elemt[1]

  elif factor_centaine == 4 :
    cent_part = first_elemt[1]+second_elemt[1]

  elif 4 < factor_centaine < 9 :
    new_factor = int(second_elemt[0]/first_elemt[0])
    fact_diff = factor_centaine - new_factor
    cent_part = second_elemt[1]+fact_diff*first_elemt[1]

  elif factor_centaine == 9 :
    cent_part = first_elemt[1]+third_elemt[1]
  
  return cent_part


# ====================================================


def roman_number_convertor(input):
  base_numbers = [1, 5, 10, 50, 100, 500, 1000]
  roman_base_numbers = ["I", "V", "X", "L", "C", "D", "M"]
  number_splitted, factors = [], []

  input = input

  mil_part, cent_part, diz_part, unit_part, result = "", "", "", "", ""

  b = input
  divider = 1000

  for i in range(4):
    mil_fac = int(b/divider)
    factors.append(mil_fac)
    a = mil_fac*divider
    number_splitted.append(int(a))
    b = b - a
    divider /= 10

  # print(factors)
  # print(number_splitted)


  # TRAITEMENT 1000

  index_1000 = base_numbers.index(1000)
  mil_part =  factors[0]*roman_base_numbers[index_1000]

  # TRAITEMENT 100
  cent_part = decomp(100, base_numbers, factors, 1, roman_base_numbers)

  # TRAITEMENT 10
  diz_part = decomp(10, base_numbers, factors, 2, roman_base_numbers)

  # TRAITEMENT 1
  unit_part = decomp(1, base_numbers, factors, 3, roman_base_numbers)

  result = mil_part + cent_part + diz_part + unit_part
  # print(result)
  return result
```
